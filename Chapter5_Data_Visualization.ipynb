```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
# tells Jupyter to display Matplotlib plots directly in the output, without having to save plots as image files or open them in a separate window.
```

```python
#One dimensional Data Set
np.random.seed(1000)
y = np.random.standard_normal(20) # generate 20 standard normally distributed (pseudo)random numbers
x = range(len(y))  # specify a list or an array for x, y coordinates
plt.plot(x, y)

#if only provide y values, plot takes the index value as the repective x values, it generates the same output
plt.plot(y)
```

```python
# pass the object with a method attached
# cumsum: the cumulative sum of this data
plt.plot(y.cumsum())
```

```python
# customize the font, labels, grid...
plt.plot(y.cumsum())
plt.grid(True) # adds a grid
plt.axis('tight') # adjusts the axis ranges
```

```python
plt.plot(y.cumsum())
plt.grid(True)
plt.xlim(-1, 20)  # directly set the minimum and maximum values for each axis
plt.ylim(np.min(y.cumsum()) - 1,
          np.max(y.cumsum()) + 1)
```

```python
plt.figure(figsize=(7, 4))
 # the figsize parameter defines the size of the figure in (width, height)
plt.plot(y.cumsum(), 'b', lw=1.5) # assign a color
plt.plot(y.cumsum(), 'ro')  # assign a symbol
plt.grid(True)
plt.axis('tight')
plt.xlabel('index') # define the labels
plt.ylabel('value')
plt.title('A Simple Plot')
```

```python
# Two-dimensional Dataset
np.random.seed(2000)
y = np.random.standard_normal((20, 2)).cumsum(axis=0)
```

```python
plt.figure(figsize=(7, 4))
plt.plot(y[:, 0], lw=1.5, label='1st')  # plots two lines
plt.plot(y[:, 1], lw=1.5, label='2nd')
plt.plot(y, 'ro')  # plots two dotted lines
plt.grid(True)
plt.legend(loc=0)  #0: best possible
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot')
```

```python
#scaling

y[:, 0] = y[:, 0] * 100
plt.figure(figsize=(7, 4))
plt.plot(y[:, 0], lw=1.5, label='1st')
plt.plot(y[:, 1], lw=1.5, label='2nd')
plt.plot(y, 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot')
```

```python
# when "visually lost" situation occrued like the chart above, There are two basic approaches to resolve this problem

# 1. Use of two y-axes (left/right)

fig, ax1 = plt.subplots()  # plot first data set using first (left) axis
plt.plot(y[:, 0], 'b', lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')
plt.grid(True)
plt.legend(loc=8)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value 1st')
plt.title('A Simple Plot')
ax2 = ax1.twinx()  # plot second data set using second (right) axis
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(y[:, 1], 'ro')
plt.legend(loc=0)
plt.ylabel('value 2nd')
```

```python
# 2. Use of two subplots (upper/lower, left/right):  generate a second subplot that shares the x-axis with the first subplot
plt.figure(figsize=(7, 5))
plt.subplot(211)
plt.plot(y[:, 0], lw=1.5, label='1st')
plt.plot(y[:, 0], 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.ylabel('value')
plt.title('A Simple Plot')
plt.subplot(212)
plt.plot(y[:, 1], 'g', lw=1.5, label='2nd')
plt.plot(y[:, 1], 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
```

```python
# scatter plot

y = np.random.standard_normal((1000, 2))

plt.figure(figsize=(7, 5))
plt.plot(y[:, 0], y[:, 1], 'ro')  # or plt.scatter(y[:, 0], y[:, 1], marker='o')

plt.grid(True)
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter Plot')
```

```python
# mathematically inspired plot (include mathematical formula in the plot)
from matplotlib.patches import Polygon

#define the function to be integrated
def func(x):
  return 0.5 * np.exp(x) + 1

# integral limits and
a, b = 0.5, 1.5 # integral limits the generation of needed numerical values
x = np.linspace(0, 2)
y = func(x)

# plot the function itself
fig, ax = plt.subplots(figsize=(7, 5))
plt.plot(x, y, 'b', linewidth=2)
plt.ylim(ymin=0)

# Illustrate the integral value, i.e. the shaded area under the function
# between the lower and upper limits
Ix = np.linspace(a, b)
Iy = func(Ix)
verts = [(a, 0)] + list(zip(Ix, Iy)) + [(b, 0)]
poly = Polygon(verts, facecolor='0.7', edgecolor='0.5')
ax.add_patch(poly)

# addition of the mathematical formula
plt.text(0.5 * (a + b), 1, r"$\int_a^b f(x)\mathrm{d}x$",
        horizontalalignment='center', fontsize=20)

plt.figtext(0.9, 0.075, '$x$')
plt.figtext(0.075, 0.9, '$f(x)$')

#set the individual x and y tick labels at their respective positions
ax.set_xticks((a, b))
ax.set_xticklabels(('$a$', '$b$'))
ax.set_yticks([func(a), func(b)])
ax.set_yticklabels(('$f(a)$', '$f(b)$'))
plt.grid(True)
```

```python
strike = np.linspace(50, 150, 24)
ttm = np.linspace(0.5, 2.5, 24)
strike, ttm = np.meshgrid(strike, ttm)

iv = (strike - 100) ** 2 / (100 * strike) / ttm # generate fake implied volatilities
strike[:2]
```

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(9, 6))
ax = fig.add_subplot(projection='3d')
surf = ax.plot_surface(strike, ttm, iv, rstride=2, cstride=2,
cmap=plt.cm.coolwarm, linewidth=0.5,
antialiased=True)
ax.set_xlabel('strike')
ax.set_ylabel('time-to-maturity')
ax.set_zlabel('implied volatility')
fig.colorbar(surf, shrink=0.5, aspect=5)
```
